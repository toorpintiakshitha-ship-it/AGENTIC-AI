```python
topic = input("What topic do you want to study today? ")

chat = model.start_chat(history=[])

print(f"\nStudy Buddy Ready for {topic}")
print("Type 'quit' to exit.\n")

while True:
    question = input("You: ")

    if question.lower() == "quit":
        print("Good luck studying!")
        break

    prompt = f"""
    You are a friendly study buddy helping students learn {topic}.
    Explain concepts simply with examples.
    Keep answers under 150 words.
    Ask one follow-up question.

    Student question:
    {question}
    """

    response = chat.send_message(prompt)

    print("\nBuddy:", response.text)
    print()
```
